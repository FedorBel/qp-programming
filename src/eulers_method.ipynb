### Euler's Method (First Order Runge-Kutta)

Example 1: Approximation of First Order Differential Equation with No Input

Solve $y'(t) = -2y(t)$ with $y_0 = 3$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import matplotlib.pyplot as plt
import numpy as np
y0 = 3
h = 0.2
t = np.arange(0, 2, 0.05)

# Exact solution
def yexact_fun(t):
    return 3 * np.exp(-2 * t)

yexact = np.array([yexact_fun(i) for i in t])

# Initial condition gives solution at t=0

def first_order_runge_kutta(h):
    global t, y0
    ystar = np.zeros(len(t))
    ystar[0] = y0
    for i in range(len(t) - 1):
        k1 = -2 * ystar[i]
        ystar[i + 1] = ystar[i] + k1 * h
    return ystar

plt.plot(t, yexact, label='Exact')
ystar = first_order_runge_kutta(0.05)
plt.plot(t, ystar, label='h = 0.05')
ystar1 = first_order_runge_kutta(0.2)
plt.plot(t, ystar1, label='h = 0.2')
ystar = first_order_runge_kutta(0.8)
plt.plot(t, ystar, label='h = 0.8')
plt.legend(loc="upper right")
plt.xlabel('t')
plt.ylabel('y')
plt.ylim(0, 2)
plt.ylim(-1, 3)
```
